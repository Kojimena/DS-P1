```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_data = pd.read_csv('../data_csv/total_revisado.csv', sep=';')
data = original_data.copy()
data.head()
```

```python
print(data['CODIGO'].nunique() == data.shape[0])
```

```python
# Limpiar los códigos, quitando los guiones solo de la primera columna
for i in range(data.shape[0]):
    data.iloc[i, 0] = int(data.iloc[i, 0].replace('-', ''))
```

```python
data.head()
```

```python
print(data.dtypes)
```

```python
# limpiar de guiones la segunda columna
for i in range(data.shape[0]):
    d = data.iloc[i, 1]
    if type(d) == str:
        d = int(d.replace('-', ''))
    elif np.isnan(d):
        d = np.nan

    data.iloc[i, 1] = d
```

```python
data.head()
```

```python
# convertir a int la primera y segunda columna
data['CODIGO'] = data['CODIGO'].astype('int64')
```

```python
print(data['CODIGO'].nunique() == data.shape[0])
```
